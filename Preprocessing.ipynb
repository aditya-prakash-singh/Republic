```python
def convert2(obj):
    L=list(obj.split(' '))
    ll=[]
    for i in L:
        ll.append(i.lower())
    print(ll)
    return(list(set(ll)))
#function to raplace braces and make tags alphanumric only
def convert3(obj):
    s=""
    for i in obj:
        if i!=" ":
            s+=i
    return(s)
def convert4(obj):
    L=[obj]
    return(L)
```

```python
mv['tags']=mv['title'].apply(convert2)
mv['article']=mv['article'].apply(convert3)
mv['tags']+=mv['article'].apply(convert4)
```

```python
def changee(obj):
    L=list(obj.split(' '))
    ll=[]
    for i in L:
        ll.append(i.lower())
    lll=[]
    for j in ll:
        s=""
        for k in j:
            if k.isalnum():
                s+=k
        lll.append(s)
    #print(ll)
    return(list(set(lll)))
mv['tags']=mv['prompt'].apply(changee)
```
